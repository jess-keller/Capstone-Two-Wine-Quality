<b> Exploratory Data Analysis: Wine Quality <b>

EDA is performed on the red wine dataset, white wine dataset, and combined wine dataset to explore the features in the datasets, investigate the relationships between features, and understand how the features compare to the response variable (quality score).

```python
# Imports 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Import datasets (red, white, combined)

red_df = pd.read_csv('clean_wine_quality_red')
white_df = pd.read_csv('clean_wine_quality_white')
combined_df = pd.read_csv('clean_wine_quality_combined')
```

**Response variable: quality score**

The average quality score of red wine (5.636023) is lower than that of white (5.877909). The average quality score of combined red and white wines is 5.818378, which makes sense given that we have approximately 3 times as many white samples as red samples.  The range of quality scores for all wine is between 3 and 9. However, the range of quality socres for red wines is between 3  and 8. Unlike white wines, no red wines received a quality score of 9.  No wine sample received a score of 1 or 2, nor a perfect 10.  Below is the distribution of quality scores for red, white, and combined.

```python
# Distribution of quality scores - COMBINED RED AND WHITE WINES

ax = sns.countplot(data=combined_df, x='quality')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.title('Quality Score for Red & White Wines')
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+50))
plt.show()
```

```python
# Distribution of quality scores - RED WINES 

ax = sns.countplot(data=red_df, x='quality')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.title('Quality Score for Red Wines')
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+10))
plt.show()
```

```python
# Distribution of quality scores - WHITE WINES 

ax = sns.countplot(data=white_df, x='quality')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.title('Quality Score for White Wines')
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+50))
plt.show()
```

As we can see, most wines received a score of 5 or 6, and to a lesser extent 7. Scores of 3, 4, or 8 were not very common and a score of 9 was very rare (only 5 out of 6497). High scoring wines appear to skew white (175 white 8s compared to only 18 red 8s and 5 white 9s compared to 0 red 9s). The quality scores appear to be generally normally distributed.

```python
# Correlation heatmap to identify feature correlations - COMBINED RED AND WHITE WINES 

corrmat = combined_df.corr()
f, ax = plt.subplots(figsize=(12, 9))
hm = sns.heatmap(round(corrmat,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Feature Correlation for Red and White Wines', fontsize=14)
```

For combined red and white wines, there is a high correlation between free sulfur dioxide and total sulfur dioxide (0.72) and a moderate correlation between residual sugar and density (0.55). There is a moderately high negative correlation between density and alcohol content (-0.69).  There is no single feature with a strong correlation to quality, but to note, the alcohol content has the strongest correlation to quality of any single feature at 0.44 (the same as for white wines and slighlty lower than in red wines, at 0.48). 

```python
# Correlation heatmap to identify feature correlations - RED WINES 

corrmat = red_df.corr()
f, ax = plt.subplots(figsize=(12, 9))
hm = sns.heatmap(round(corrmat,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Feature Correlation for Red Wines', fontsize=14)
```

For red wines, there is a moderately high correlation between citric acid and fixed acidity (0.67) as well as between free sulfur dioxide and total sulfur dioxide (0.67). There is a moderate negative correlation between pH and fixed acidity (-0.68) and between pH and citric acid (-0.54).  There is no single feature with a strong correlation to quality, but to note, alcohol content has the strongest correlation to quality of any single feature at 0.48. 

```python
# Correlation heatmap to identify feature correlations - WHITE WINES 

corrmat = white_df.corr()
f, ax = plt.subplots(figsize=(12, 9))
hm = sns.heatmap(round(corrmat,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Feature Correlation for White Wines', fontsize=14)
```

For white wines, there is a high correlation between density and residual sugar (0.84) as well as a moderately high correlation between free sulfur dioxide and total sulfur dioxide (0.62). There is a negative correlation between density and alcohol content (-0.78).  There is no single feature with a strong correlation to quality, but to note, like with the red wines, the alcohol content has the strongest correlation to quality of any single feature at 0.44.

**Conclusions:** 

The is no single variable that strongly correlates with quality, but the variable with the highest correlation to quality is the alcohol content, which holds true for both red and white wines. It should be noted that the correlation is weak, at only 0.44 for white and combined wines and 0.48 for red wines. 

The average alcohol content for all wines was 10.491801% by volume (10.422983% by voulme for reds and 10.514267% by volume for whites), with a range of 8.0% to 14.9%.  The ranges for reds and whites was 8.4% to 14.9% and 8.0% to 14.2%, respectively. There is not great variation between the means or ranges of whites and reds, although the wines with the highest alcohol content (above 14.2% by volume) are reds.  

For combined wines, the notable correlations between variables were free sulfur dioxide and total sulfur dioxide (0.72), residual sugar and density (0.55), and density and alcohol content (-0.69).

For red wines, the notable correlations between variables were free sulfur dioxide and total sulfur dioxide (0.67), citric acid and fixed acidity (0.67), and pH and fixed acidity (-0.68).

For white wines, the notable correlations between variables were density and residual sugar: (0.84), free sulfur dioxide and total sulfur dioxide (0.62), and density and alcohol content (-0.78). 

Across all wines, there was a correlation between free sulfur dioxide and total sulfur dioxide, which makes sense as each of these variables is a measure of the same chemical compound. Per the variable description provided with the data, free sulfur dioxide measures the SO2 that exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion, while total sulfur dioxide measures the amount of free and bound forms of S02.  

Sulfur dioxide aside, variables that may be of interest are: residual sugar, density, citic acid, fixed acidity, pH, and alcohol content.  Fixed acidity, citric acid and pH are all related to acidity; residual sugar is the amount of sugar remaining after fermentation stops; and density is dependent upon the percent alcohol and sugar content. It is intuitive that these interrealted physiochemical measures may impact wine taste and quality scores.
